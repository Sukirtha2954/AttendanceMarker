```python
import pandas as pd

# Step 1: Load the Excel sheet that contains names and roll numbers
attendance_df = pd.read_excel('Copy of 03.01.2023 - MORNING.xlsx', sheet_name='AI&DS(B)')

# Step 2: Function to mark attendance for a specific date
def mark_attendance_for_date(date_column):
    print(f"Marking attendance for {date_column}:")
    
    # Initialize an empty list to store attendance status
    status_list = []
    
    # Loop through each row (roll number) and ask for attendance
    for index, row in attendance_df.iterrows():
        roll_number = row['Roll Number']
        status = input(f"Is roll number {roll_number} present? (y/n): ").strip().lower()
        if status == 'y':
            status_list.append('Present')
        else:
            status_list.append('Absent')
    
    # Add the status to the respective column in the DataFrame
    attendance_df[date_column] = status_list

# Step 3: Prompt for a date and mark attendance for that date
def take_attendance():
    while True:
        date = input("Enter the date for marking attendance (in YYYY-MM-DD format or type 'exit' to quit): ").strip()
        
        if date.lower() == 'exit':
            break
        
        if date not in attendance_df.columns:
            # Add a new column for this date if it doesn't already exist
            attendance_df[date] = None
        
        # Call the function to mark attendance for that date
        mark_attendance_for_date(date)
        
        # Ask if you want to continue marking for another date
        continue_marking = input("Do you want to mark attendance for another date? (y/n): ").strip().lower()
        if continue_marking != 'y':
            break

# Step 4: Save the updated Excel sheet
def save_attendance():
    attendance_df.to_excel('Copy of 03.01.2023 - MORNING.xlsx', index=False)
    print("Attendance updated and saved.")

# Example usage:
# Call this to take attendance for multiple dates until you type 'exit'
take_attendance()

# After marking attendance for the desired dates, save the changes
save_attendance()
```

```python
import pandas as pd

# Step 1: Load the Excel sheet with attendance data
attendance_df = pd.read_excel('Copy of 03.01.2023 - MORNING.xlsx', sheet_name='Sheet1')

# Step 2: Define the list of dates for Mondays and Tuesdays
monday_dates = ['2024-08-12', '2024-08-19', '2024-09-02', '2024-09-09', '2024-09-16']
tuesday_dates = ['2024-08-13', '2024-08-20', '2024-08-27', '2024-09-03', '2024-09-10']

# Step 3: Define a function to calculate attendance points based on these dates
def calculate_attendance_with_custom_points():
    # Initialize a list to store the total points for each student
    total_points_list = []
    
    # Loop through each student and calculate their total points
    for index, row in attendance_df.iterrows():
        total_points = 0
        
        # Loop through each date column to check attendance and assign points
        for date_col in attendance_df.columns:
            if row[date_col] == 'Present':
                # Assign points based on the specific dates
                if date_col in monday_dates:
                    total_points += 2  # Monday earns 2 points
                elif date_col in tuesday_dates:
                    total_points += 3  # Tuesday earns 3 points
                else:
                    total_points += 1  # Default 1 point for other days
        
        # Append the total points for the student
        total_points_list.append(total_points)
    
    # Add the total points as a new column
    attendance_df['Total Points'] = total_points_list

# Step 4: Save the updated Excel sheet with the calculated total points
def save_updated_sheet():
    attendance_df.to_excel('Copy of 03.01.2023 - MORNING_UPDATED.xlsx', index=False)
    print("Attendance points calculated and saved.")

# Example usage:
# Call the function to calculate attendance points
calculate_attendance_with_custom_points()

# Save the updated Excel file with attendance points
save_updated_sheet()
```
